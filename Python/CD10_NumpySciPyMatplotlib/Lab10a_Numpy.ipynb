## Phần 1: Array Basics và Indexing

### Bài 1.1: Tạo và thao tác với mảng

```python
import numpy as np

# TODO: Tạo mảng 1D chứa các số từ 0 đến 99
arr_1d = np.arange(100)
# dùng kỹ thuật slicing để in 10 phần tử đầu tiên
print("Mang 1D:", arr_1d[:10], "...")  # In 10 phần tử đầu tiên (in ra phần tử từ 0 - 9)

# TODO: Reshape mảng trên thành mảng 2D kích thước 10x10
arr_2d = arr_1d.reshape(10, 10)

# TODO: Lấy phần tử ở hàng 3, cột 5
element = arr_2d[3, 5]

# TODO: Lấy hàng thứ 5
#dùng kỹ thuật slicing để lấy toàn bộ phần tử trong hàng 5
row_5 = arr_2d[5, :] #hoặc arr_2d[5]

# TODO: Lấy cột thứ 7
col_7 = arr_2d[:, 7] #tất cả các hàng , cột 7

# TODO: Lấy subarray 3x3 từ góc trên bên trái
subarray = arr_2d[0:3, 0:3] # hàng lấy từ 0 đến 3-1, cột lấy từ 0 đến 3-1

print("Mảng 2D:\n", arr_2d)
print("\nPhần tử [3,5]:", element)
print("\nHàng 5:", row_5)
print("\nCột 7:", col_7)
print("\nSubarray 3x3:\n", subarray)
```

- **np.arange** là hàm NumPy dùng để tạo mảng 1D chứa các số theo dãy số họ có khoảng cách đều nhau

- **reshape()** biến đổi hình dạng (shape) của mảng từ 1 chiều thành nhiều chiều mà **Không làm thay đổi dữ liệu gốc**

- câu lệnh **indexing[a,b]** dùng để truy cập phần tử cụ thể trng mảng đa chiều bằng chỉ 
- câu lệnh **Slicing hàng[a]** lấy toàn bộ hàng từ mảng a(2D)
- câu lệnh **Slicing cột[:, a] hoặc [a]** lấy toàn bộ cột a (2D)

### Bài 1.2: Boolean Indexing

```python
# Tạo mảng ngẫu nhiên
np.random.seed(42)
data = np.random.randn(100)

# TODO: Tìm tất cả các phần tử > 0
positive_elements = data[data > 0] #data > 0: tạo một mảng boolean với true ở các vi trí phần tử >0
#data[data >0] sử dụng boolean indexing để lọc ra chỉ những phần tử thoả mãn điều kiện

# TODO: Tìm tất cả các phần tử trong khoảng [-1, 1]
range_elements = data[(data >= -1) & (data <= 1)] #sử dụng toán tử & để kết hợp hai điều kiện boolean

# TODO: Thay thế tất cả các phần tử âm bằng 0
data_modified = data.copy() # tạo bản sao để giữ nguyên mảng gốc
data_modified[data_modified < 0] = 0 #gán giá trị 0 cho các phần tử âm sử dụng boolean indexing

print("Số phần tử dương:", len(positive_elements))
print("Số phần tử trong [-1,1]:", len(range_elements))
print("Số phần tử âm đã thay thế:", np.sum(data < 0))
```

### Bài 2.2: Broadcasting

```python
# Tạo ma trận và vector
matrix = np.random.randn(4, 3)
vector = np.array([1, 2, 3])

print("Ma trận (4x3):\n", matrix)
print("\nVector (3,):", vector)

# TODO: Cộng vector vào mỗi hàng của ma trận sử dụng broadcasting
result_broadcast = matrix + vector  # broadcasting tự động mở rộng vector

# TODO: Nhân mỗi cột của ma trận với [1, 2, 3, 4] sử dụng broadcasting
# Hint: Cần reshape vector thành (4, 1)
col_multiplier = np.array([1, 2, 3, 4]).reshape(4, 1)
result_col = matrix * col_multiplier
print("\nNhan theo cot:\n", result_col)

# TODO: Chuẩn hóa ma trận (normalize) - trừ mean và chia cho std theo mỗi cột
normalized_matrix = np.mean(matrix, axis=0)
std_col = matrix * col_multiplier
std_col = np.std(matrix, axis=0)
normalized_matrix = (matrix - normalized_matrix) / std_col
print("\nMa trận chuẩn hóa:\n", normalized_matrix)


print("\nKết quả broadcast:\n", result_broadcast)
print("\nNhân theo cột:\n", result_col)
print("\nMa trận chuẩn hóa:\n", normalized_matrix)
```
